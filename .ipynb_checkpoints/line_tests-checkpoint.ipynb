```python
# Donald Nguyen
# Created 09/28/2021

# Imports
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from astropy.io import fits

%matplotlib inline
```

```python
from fits_handler import FitsFileHandler

handler = FitsFileHandler()
#data = handler.data_from_single_fits(handler.filepaths[0], 1)
datas = [handler.data_from_single_fits(filepath, 1) for filepath in handler.filepaths]

barys = [data['bary_wavelength'] for data in datas]
continuums = [data['continuum'] for data in datas]
spectra = [data['spectrum'] for data in datas]

obs_ex = []
order_ex = 47
obs_range_ex = (3500, 4000)
for i in range(len(barys)):
    obs_ex.append(spectra[i][order_ex, obs_range_ex[0]:obs_range_ex[1]] / continuums[i][order_ex, obs_range_ex[0]:obs_range_ex[1]])

plt.figure(figsize=(16, 9))
for bary, ob in zip(barys, obs_ex):
    plt.plot(bary[order_ex, obs_range_ex[0]:obs_range_ex[1]], ob)
plt.vlines(5420.35, 0, 1)
plt.show()

```

```python
from line_list_loader import LineListLoader_Wise

@np.vectorize
def wavelength_air2vac(wavelength_air): 
    '''Wavelength in vacuum from wavelength in air in units of Angstroms'''
    s = 1e4 / wavelength_air
    n = 1 + 0.00008336624212083 + 0.02408926869968 / (130.1065924522 - s*s) + 0.0001599740894897 / (38.92568793293 - s*s)
    return wavelength_air * n

min_VALD_depth = 0.5
max_VALD_depth = 0.8

wise_lines = LineListLoader_Wise()
chosen_lines = wise_lines.list_VALD_depths(min_VALD_depth, max_VALD_depth)
print(chosen_lines)


obs = [spectra[i] / continuums[i] for i in range(len(barys))]

chosen_lines_positions = {}

for chosen_line in chosen_lines['Wavelength']: # Wavelengths in air
    chosen_lines_positions[chosen_line] = []
    for ind_file, (bary, ob) in enumerate(zip(barys, obs)):
        for order in np.arange(len(bary)):
            if bary[order, 0] <= wavelength_air2vac(chosen_line) <= bary[order, -1]:
                chosen_lines_positions[chosen_line].append((ind_file, order))
```

```python

fig, axes = plt.subplots(chosen_lines['Wavelength'].count(), 1, figsize =(16, 20))
plt.tight_layout()
wave_range = 20
for ax, chosen_line, positions in zip(axes, chosen_lines['Wavelength'], chosen_lines_positions.values()):
    chosen_line_vac = wavelength_air2vac(chosen_line)
    ax.vlines(chosen_line_vac, 0, 1)
    ax.set_xlim(chosen_line_vac - wave_range / 2, chosen_line_vac + wave_range / 2)
    for pos in positions:
        ax.plot(barys[pos[0]][pos[1]], obs[pos[0]][pos[1]])
plt.show()

# TODO
# bary_wavelengths are prob wrt sun, need to add earth rv shift?
```
